```python
import random


def fitness(population, score, target):
  fit_scale = []
  ret_index = 0
  for sample in population:
    sum = 0
    for i in range(len(score)):
      if sample[i] == '1':
        sum += score[i]


    fit_scale.append(abs(target - sum))
    if sum == target:
      return ret_index, fit_scale
    ret_index += 1




  return -1, fit_scale








def select(population, fit_scale):
  #sorting depending on fit_scale
  for i in range(len(fit_scale)-1):
      temp_fit = fit_scale[i+1]
      temp_pop = population[i + 1]
      j = i

      # descending order
      while j >= 0:
          if fit_scale[j] < temp_fit:
              fit_scale[j+1] = fit_scale[j]
              population[j + 1] = population[j]
          else:
              break

          j -= 1

      fit_scale[j+1] = temp_fit
      population[j + 1] = temp_pop


  return population






def crossover(x, y):
  mid = len(x)//2


  x1 = x[0:mid]
  x2 = x[mid:len(x)]


  y1 = y[0:mid]
  y2 = y[mid:len(y)]


  x = x1 + y2
  y = y1 + x2

  return x, y








def mutate(child):
  index = random.randint(0, len(child)-1)


  ret = ""
  if child[index] == '0':
    if index == len(child) -1:
      ret = child[0:index] + '1'
    else:
      ret = child[0:index] + '1' + child[index+1:len(child)]
  else:
    if index == len(child) -1:
      ret = child[0:index] + '0'
    else:
      ret = child[0:index] + '0' + child[index+1:len(child)]



  return ret










def GA(population, n, target):
  cnt = 0
  while cnt < 1000:
    ans, fit_scale  = fitness(population, score, target)


    if ans != -1:
      return population[ans]


    else:
      population = select(population, fit_scale)

      new_population = []
      for i in range(0, len(population), 2):
        p1, p2 = crossover(population[i], population[i+1])
        new_population.append(p1)
        new_population.append(p2)


      for i in range(len(new_population)):
        new_population[i] = mutate(new_population[i])


      population = new_population

    cnt += 1
  return -1







input = open("/content/input.txt")


n, t = (input.readline().strip()).split(' ')
n = int(n)
t = int(t)


players = [0] * n
score = [0] * n


for i in range(n):
  players[i], score[i]= input.readline().strip().split()
  score[i] = int(score[i])


population = []
for i in range(10):
  random_population = ""
  for j in range(n):
    random_population += str(random.randint(0,1))
  population.append(random_population)


ans = GA(population, n, t)


print(players)
print(ans)



```
